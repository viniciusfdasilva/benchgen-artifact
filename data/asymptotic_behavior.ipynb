```python
import seaborn as sns, pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import MinMaxScaler
```

```python
dtype = {'clang_time': float,'opt_time': float,'llc_time': float, 'bin_size': int,'opt': str, 'iteration': int, 'grammar_name': str, 'data_structure': str}
dataframe = pd.read_csv('asymptotic_behavior.csv', sep=',', dtype=dtype)
```

```python
dataframe['bin_size_in_kb'] = dataframe['bin_size']/1024 
```

```python
opts = ['-O0', '-O1', '-O2', '-O3', '-O3 -ffast-math']
cmap = ListedColormap(sns.color_palette("husl", len(opts)).as_hex())
```

### Data Structure: Array
### Program Id: ex6
### Program:
<html>
A = IF(A, new new new);</br>
</html>

```python
array_ex6 = dataframe[(dataframe['data_structure'] == 'array') & (dataframe['grammar_name'] == 'ex6')]
```

```python
scaler = MinMaxScaler()

df = array_ex6.copy()
df['opt_time_norm'] = scaler.fit_transform(df[['opt_time']])

plt.figure(figsize=(7, 6), dpi=150)
sns.lineplot(data=df, x='bin_size_in_kb', y='opt_time_norm', hue='opt', style='opt',errorbar=None, markers=True, dashes=False, linewidth=2)

plt.xlabel("Binary size (KB)")
plt.ylabel("Clang-opt time (s)")
plt.grid(True)

handles = [plt.Line2D([0], [0], marker='o', color='w',
           markerfacecolor=cmap(i), markersize=10,
           label=level)
          for i, level in enumerate(opts)]

spec_text = """Specification of Automatically Generated Program:

Data Structure: Array

Grammar:
A = IF(A, new new new);
"""

plt.figtext(0, -0.2, spec_text, ha='left', va='bottom', fontsize=9,
            bbox={'facecolor':'white', 'alpha':0.8, 'pad':1})

plt.tight_layout()
plt.show()
```
